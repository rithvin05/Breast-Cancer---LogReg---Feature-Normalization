# Mean Logistic Regression Model Accuracy: 97.7% 

```python
df = pd.read_csv('wdbc.data')
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df.isnull().sum() + df.isna().sum()
```

```python
df['Diagnosis'] = df['Diagnosis'].map({'M':1,'B':0})
df['Diagnosis']
```

```python
X = df.iloc[:, 2:]
X
```

```python
y = df['Diagnosis']
y
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline

st_pipeline = Pipeline([('scalar', StandardScaler()),
                     ('classifier', LogisticRegression())])

min_max_pipeline = Pipeline([('scalar', MinMaxScaler()),
                     ('classifier', LogisticRegression())])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=42)

st_pipeline.fit(X_train, y_train)
st_pipeline.score(X_test, y_test)
```

```python
min_max_pipeline.fit(X_train, y_train)
min_max_pipeline.score(X_test, y_test)
```
